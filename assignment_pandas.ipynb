**4-5. Рассчитайте среднее значение и среднеквадратичное отклонение возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год**

```python
data['salary'].value_counts()

high_income=data[data['salary']=='>50K']['age']
low_income=data[data['salary']=='<=50K']['age']

print("Средний возраст (>50K):", high_income.mean())
print("СКО возраста (>50K):", high_income.std())

print("\nСредний возраст (<=50K):", low_income.mean())
print("СКО возраста (<=50K):", low_income.std())
```

**6. Правда ли, что люди, которые получают больше 50k, имеют минимум высшее образование? (признак *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

```python
higher_edu=['Bachelors','Prof-school','Assoc-acdm','Assoc-voc','Masters','Doctorate']
high_salary=data[data['salary']=='>50K']
non_higher=high_salary[~high_salary['education'].isin(higher_edu)]

if len(non_higher)==0:
	print("Да, все люди с доходом выше 50К имеют высшее образование")
else:
	print("Нет, есть люди с доходом >50K без высшего образования. Их количество:", len(non_higher))
```

**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Asian-Pac-Islander*.**

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
max_hours=data['hours-per-week'].max()
print("Максимальное число часов в неделю:", max_hours)
max_hours_workers=data[data['hours-per-week'] == max_hours]
num_max_hours=len(max_hours_workers)
print("Количество людей, работающих максимум часов:", num_max_hours)
count_hard_workers=max_hours_workers.groupby('salary')['salary'].count()
share_high_income=count_hard_workers.get(">50K",0)/num_max_hours*100
print(f'Процент зарабатывающих много: {share_high_income:.2f}%')
```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**

```python
data['hours-per-week'].mean()
avg_hours=data.groupby(['native-country', 'salary'])['hours-per-week'].mean().reset_index(name='avg_hours')
print("Cреднее время работы зарабатывающих мало и много для каждой страны:\n", avg_hours)
```

**11.Сгруппируйте людей по возрастным группам *young*, *adult*, *retiree*, где:**
* *young* соответствует 16-35 лет
* *adult* - 35-70 лет
* *retiree* - 70-100 лет

**Проставьте название соответсвтуещей группы для каждого человека в новой колонке AgeGroup**

```python
bins=[16, 35, 70, 100]
labels=['young','adult','retiree']
data['AgeGroup']=pd.cut(data['age'],bins=bins,labels=labels,right=False)
data[['age','AgeGroup']]
```

**12-13. Определите количество зарабатывающих >50K в каждой из возрастных групп (колонка AgeGroup), а также выведите название возрастной группы, в которой чаще зарабатывают больше 50К (>50K)**

```python
num_high_income_group=data[data['salary']=='>50K'].groupby('AgeGroup',observed=True).size()
print(num_high_income_group)
print('Возрастная группа, в которой чаще зарабатывают больше 50К:', num_high_income_group.idxmax())
```

**14. Сгруппируйте людей по типу занятости (колонка occupation) и определите количество людей в каждой группе. После чего напишите функцию фильтрации filter_func, которая будет возвращать только те группы, в которых средний возраст (колонка age) не больше 40 и в которых все работники отрабатывают более 5 часов в неделю (колонка hours-per-week)**

```python
# Variant 1
# print('Количество людей в каждой группе:\n', data.groupby('occupation').size().reset_index(name="num_people"))

# def filter_func(group):
# 	return (group['age'].mean()<=40) and (group['hours-per-week'].min()>5)

# filtered=data.groupby('occupation').filter(filter_func)
# counts_filtered = filtered.groupby('occupation').size().reset_index(name="num_people")
# print('\nФильтрация групп:\n', counts_filtered)

# Variant 2
data_clean=data[data['occupation']!="?"]
counts_before = data_clean.groupby('occupation').size().reset_index(name="num_people")

# агрегаты для каждой професии
agg = data_clean.groupby('occupation').agg(
    avg_age=('age', 'mean'),
    min_hours=('hours-per-week', 'min'),
    num_people=('age', 'size')
).reset_index()
agg

def filter_func(group_agg):
    return (group_agg['avg_age'] <= 40) & (group_agg['min_hours'] > 5)

mask = filter_func(agg)
agg_filtered = agg[mask]

# профессии, которые прошли фильтр
passed_occupations = agg_filtered['occupation']
counts_before['passed_filter'] = counts_before['occupation'].isin(passed_occupations)

print("\nСравнение количества людей по профессиям и фильтру:\n", counts_before)
```
