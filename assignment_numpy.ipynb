```python
arr = np.array([[1, 2],
                [3, 4]])

# строки для комбинаций
a = arr[0]  # первая строка
b = arr[1]  # вторая строка

# сетка комбинаций
A, B = np.meshgrid(a, b)

# преобразование сетки в список пар
result = np.column_stack([A.ravel(), B.ravel()])

print("Декартово произведение:\n", result)
```

* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ 3.

```python
################### PYTHON ######################

list_x = [[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]
print(list_x)
list_prod=1
for i in range(min(len(list_x),len(list_x[0]))):
	if list_x[i][i]!=0:
		list_prod*=list_x[i][i]
print("Произведение ненулевых элементов на python: ", list_prod)

################### NUMPY ######################

array_x=np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])
print(array_x)
array_prod=np.prod(np.diag(array_x)[np.diag(array_x)!=0])
print("Произведение ненулевых элементов на numpy: ", array_prod)
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
################### PYTHON ######################
list_x=[2, 2, 2, 3, 3, 3, 5]
values_py=[]
counts_py=[]
count=1
current=list_x[0]

for i in range(1, len(list_x)):
	if list_x[i]==current:
		count+=1
	else:
		values_py.append(current)
		counts_py.append(count)
		count=1
		current=list_x[i]

values_py.append(current)
counts_py.append(count)

print("Python Числа: ",values_py)
print("Python Количество: ",counts_py)

################### NUMPY ######################

array_x = np.array([2, 2, 2, 3, 3, 3, 5])
diff_x=np.diff(array_x)
change_idx=np.flatnonzero(diff_x)+1
idx_boundaries = np.r_[0,change_idx,len(array_x)]
values_np=array_x[idx_boundaries[:-1]]
counts_np=np.diff(idx_boundaries)

print("Numpy Числа:", values_np)
print("Numpy Количество:", counts_np)
```

* __Задача 5__: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Сравните с функцией scipy.spatial.distance.cdist по скорости работы.

```python
import math
import time
from scipy.spatial.distance import cdist

################### PYTHON ######################

list_x=[[1,2], [3,4]]
list_y=[[0,0], [1,1], [5,5]]

#Python
start=time.time()
euclid_dist_py=[]
for x in list_x:
	row_dist=[]
	for y in list_y:
		row_dist.append(math.sqrt(sum([(xi-yi)**2 for xi, yi in zip(x,y)])))
	euclid_dist_py.append(row_dist)
end=time.time()
print("\nPython time:", end-start)
print("Матрица расстояний (Python):\n", euclid_dist_py)

################### NUMPY ######################

array_x=np.array([[1,2], [3,4]])
array_y=np.array([[0,0], [1,1], [5,5]])

start=time.time()
diff=array_x[:,np.newaxis,:]-array_y[np.newaxis, :, :]
euclid_dist_np=np.sqrt(np.sum(diff**2,axis=2))
end=time.time()
print("\nNumPy time:", end - start)
print("Матрица расстояний (NumPy):\n", euclid_dist_np)

################### SCIPY ######################

start=time.time()
euclid_dist_scipy=cdist(list_x,list_y,metric='euclidean')
end=time.time()
print("\nSciPy time:", end-start)
print("Матрица расстояний (SciPy):\n", euclid_dist_scipy)
```
